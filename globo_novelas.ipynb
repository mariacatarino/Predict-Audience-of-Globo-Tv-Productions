```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.tree import DecisionTreeRegressor
```

```python
dataset = pd.read_excel("GloboNovelas.xlsx")
```

```python
dataset
```

```python
dataset.to_csv("GloboNovelas.csv",index=None,header=True)
```

```python
dataset_csv = pd.read_csv("GloboNovelas.csv")
```

```python
dataset_csv
```

```python
dataset_csv = dataset_csv.drop(['Inicio','Final','Novela','Faixa','Maiores','Mais Famosas','Classificação','Menu'],axis=1)
```

```python
dataset_csv
```

```python
dataset_csv = pd.get_dummies(dataset_csv, columns=['Autor'])
```

```python
dataset_csv
```

```python
y = dataset_csv.Audiencia
```

```python
X = dataset_csv.drop(['Audiencia'],axis=1)
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
```

```python
dataset_model = DecisionTreeRegressor()
```

```python
dataset_model.fit(train_X, train_y)
```

```python
val_predictions = dataset_model.predict(val_X)
```

```python
print("Making predictions for the following exemples:")
print(val_y)
print("The predictions are")
print(val_predictions)
```

```python
r2_score(val_predictions,val_y)
```
